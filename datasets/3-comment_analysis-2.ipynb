```python
import pandas as pd;
#timestamps = pd.read_csv("answer_timestamps.csv")
#timestamps.set_index('Timestamp')
#timestamps.index = pd.to_datetime(timestamps["Timestamp"])
comments = pd.read_csv("answer_comments.csv")
```

```python
#let's count and plot answers based on a comment prevalence
#make a function for this to make it more DRY
def print_comment_summary( column, plot_title ):
    comment_answers = comments.loc[comments[column] == True]
    print(comment_answers['Q1 CLASSIFICATION'].value_counts())
    print("---")
    print(comment_answers["Q2 CLASSIFICATION"].value_counts(normalize=True)) # get percentages of various respondent categories
    print("---")
    print(comment_answers["Q2 CLASSIFICATION"].value_counts().plot(kind="pie", title=plot_title)) # print a pie to make comparison to the main chart
```

```python
print_comment_summary("COMMENT(INEFFICIENT_MANAGEMENT_HIRING_AND_WORK_PRACTICES)", "Mentions issues in management, hiring or work practices leading to need for more labor")
```

```python
print(comments.loc[comments["COMMENT(INEFFICIENT_MANAGEMENT_HIRING_AND_WORK_PRACTICES)"] == True]["KEY_TAKEAWAYS_FROM_COMMENTS"].sample(5).values)
```
